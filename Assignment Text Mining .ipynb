# Business Analysis on Consumer Reviews of Amazon Products 

```python
import pandas as pd 
import numpy as np
import seaborn as sns
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
#Reading csv file
az = pd.read_csv('F:/ipba/Text Mining/Consumer_Reviews_of_Amazon_Products.csv')
```

### Customer Review rating on different Product Categories 

```python
# Plotting graph (answer for question 2 and 3)
plt.figure(figsize=(15,5))
chart = sns.countplot(x='primaryCategories',hue ='reviews_rating',data = az)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

### Part Of Speech (POS)  tagging for the word "seems" in negative feedback 

```python
import spacy
from spacy import displacy
from collections import Counter
import en_core_web_sm    
```

```python
nlp = en_core_web_sm.load()
```

```python
import nltk 
from nltk import word_tokenize,pos_tag
```

```python
neg_review_rating = az.loc[az['reviews_rating']== 1]
    
```

```python
neg_review_rating[:5]
```

```python
neg_word = []
for i in neg_review_rating.reviews_text:
    neg_word.append(i)
neg_word[:4]
```

```python
for t in range(0,2):
    print(pos_tag(word_tokenize(neg_word[t])))
    print("\n")
```

## WordCloud for negative feedback (Review Rating 1)  

```python
#for negative comments 
neg_text = pd.Series(neg_word).str.cat(sep = ' ')
```

```python
from wordcloud import WordCloud
```

```python
from wordcloud import STOPWORDS
stop_words = list(STOPWORDS)
wordcloud = WordCloud(width = 1800,height = 900, max_font_size = 200).generate(neg_text)
```

```python
plt.figure(figsize = (15,15))
plt.imshow(wordcloud)
plt.axis('off')
```
